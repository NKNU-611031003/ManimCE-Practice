## Logo

```python
from manim import *
def ManimCELogo(self):
    # logo
    banner = ManimBanner(dark_theme=False)
    circle_banner = Circle(color="#ece6e2", fill_opacity=1).surround(banner).scale(2)
    # logo Text
    logo_Text = Text("Manim Community", font_size=48).shift(RIGHT)
    logo_MadeBy = Text("Made by", font_size=24).next_to(logo_Text, UP).shift(LEFT*2.5)
    logo_Web = Text("https://docs.manim.community/en/stable/index.html", font_size=16).next_to(logo_Text, DOWN)
    # Group
    banner_Group = VGroup(circle_banner, banner)
    logo_Text_Group = Group(logo_MadeBy, logo_Web)

    self.play(FadeIn(circle_banner), banner.create())
    self.play(banner.animate.scale(0.2), circle_banner.animate.scale(0.1))
    self.play(banner_Group.animate.next_to(logo_Text_Group, LEFT))
    self.play(FadeIn(logo_Text_Group), Write(logo_Text), run_time=2.5)
    self.wait(3)
    self.play(FadeOut(banner_Group), FadeOut(logo_Text_Group), FadeOut(logo_Text))
    self.wait()
```

## Question

```python
class main_polygon(object):
  def __init__(self, args):
    self.font_size = args.font_size
    self.side_length = args.side_length
    self.label_font_size = args.label_font_size
    self.dot_scale = args.dot_scale

    self.dot = list(range(6))
    self.label = list(range(6))
    self.dot_label = list(range(6))
    self.side = list(range(4))
    self.length_label = list(range(4))

    ### Set graph position ###
    self.square = Square(side_length=self.side_length*2)

  ### Draw arc graph ###
  def Arc_Draw(self):
    arc_1 = Arc(radius=self.side_length, angle=PI, arc_center=self.square.get_bottom())
    arc_2 = Arc(radius=self.side_length*2, angle=PI/2, arc_center=self.square.get_top(), start_angle=PI*3/2).shift(LEFT*self.side_length)
    return VGroup(arc_1, arc_2)

  ### Draw Dot and Label Dot ###
  def Dot_Label(self):
    def Dot_and_Label(num, name, dot_pos, label_pos):
      dot = Dot(point=self.square.get_center()).scale(self.dot_scale).shift((dot_pos)*self.side_length)
      label = Tex(name, font_size=self.label_font_size).next_to(dot, label_pos* 0.6)
      return dot, label, VGroup(dot, label)
    self.dot[0], self.label[0], self.dot_label[0] = Dot_and_Label(0, "$A$", LEFT+UP, (LEFT+UP))
    self.dot[1], self.label[1], self.dot_label[1] = Dot_and_Label(1, "$B$", RIGHT+UP, (RIGHT+UP))
    self.dot[2], self.label[2], self.dot_label[2] = Dot_and_Label(2, "$C$", RIGHT+DOWN, (RIGHT+DOWN))
    self.dot[3], self.label[3], self.dot_label[3] = Dot_and_Label(3, "$D$", LEFT+DOWN, (LEFT+DOWN))
    self.dot[4], self.label[4], self.dot_label[4] = Dot_and_Label(4, "$M$", DOWN, (DOWN)*0.8)
    self.dot[5], self.label[5], self.dot_label[5] = Dot_and_Label(5, "$P$", RIGHT*0.6+DOWN*0.2, (UP))
    return VGroup(*[self.dot[i] for i in range(6)])

  ### Draw side ###
  def Side_Draw(self):
    for i in range(3):
      self.side[i] = Line(self.dot[i], self.dot[i+1])
    self.side[3] = Line(self.dot[3], self.dot[0]) # Make AD Line
    return VGroup(*[self.side[i] for i in range(4)])

  ### Label side ###
  def Side_Label(self):
    def Side_label(num, label_name, dot_0, dot_1, direction):
      length_label_Text = Tex(label_name, font_size=self.label_font_size).next_to(Line(dot_0, dot_1).get_center(), direction)
      return length_label_Text
    self.length_label[0] = Side_label(0, "4", self.dot[0], self.dot[1], UP)
    self.length_label[1] = Side_label(1, "4", self.dot[1], self.dot[2], RIGHT)
    self.length_label[2] = Side_label(2, "2", self.dot[2], self.dot[4], DOWN)
    self.length_label[3] = Side_label(3, "2", self.dot[3], self.dot[4], DOWN)
    return VGroup(*[self.length_label[i] for i in range(4)])

  ### All graph group (except square)###
  def Graphics_Group(self):
    arc = self.Arc_Draw()
    dot_group = self.Dot_Label()
    side_draw = self.Side_Draw()
    length_label_group = self.Side_Label()
    return VGroup(arc, dot_group, side_draw, length_label_group)

  def Question_Text(self, graphics):
    return Tex(r' $P$  $\overline{AD}$ ', tex_template=TexTemplateLibrary.ctex, font_size=self.font_size).next_to(graphics, UP*2)
```

```python
%%manim -qm -v WARNING MathQuestion_221019
import argparse
import numpy as np
import math

parser = argparse.ArgumentParser()
parser.add_argument('-f')
parser.add_argument("--font_size", type=int, default=50)
######### graph setting #########
parser.add_argument("--label_font_size", type=int, default=50)
parser.add_argument("--side_length", type=float, default=2.0, help="Side length of half square.")
parser.add_argument("--dot_scale", type=float, default=1.0, help="Size of each point.")
######### animation control #########
parser.add_argument("--wait_long", type=float, default=2.0)
parser.add_argument("--wait_short", type=float, default=1.0)
parser.add_argument("--run_long", type=float, default=2.0)
parser.add_argument("--run_short", type=float, default=1.0)

class MathQuestion_221019(MovingCameraScene):
  def construct(self):
    args = parser.parse_args()

    # OP animation
    # ManimCELogo(self)
    # polygon Image
    p = main_polygon(args)
    graphics = p.Graphics_Group()
    graphics_arc = graphics[0]
    graphics_dot = graphics[1]
    square_side = graphics[2]
    graphics_sideLabel = graphics[3]

    question_Text = p.Question_Text(graphics)

    self.play(FadeIn(graphics), Write(question_Text), run_time=2.0)
    self.wait(args.wait_short)
    self.play(FadeOut(question_Text), graphics.animate.move_to(ORIGIN), run_time=2.0)
    # self.wait(args.wait_short)
    self.play(self.camera.frame.animate.move_to(LEFT*3))

    line_AP = Line(graphics_dot[0], graphics_dot[5])
    line_AP_m = line_AP.copy().set_color(YELLOW)
    line_AD_m = Line(graphics_dot[0], graphics_dot[3]).set_color(YELLOW)
    line_AD_Label = Tex("4", font_size=args.label_font_size).set_color(YELLOW).next_to(line_AD_m, LEFT)
    line_AP_Label = line_AD_Label.copy().next_to(line_AP.get_center(), (RIGHT+UP)*0.6)
    self.play(FadeIn(line_AD_m, line_AD_Label))
    self.wait(args.wait_short)
    self.play(ReplacementTransform(line_AD_m.copy(), line_AP_m, path_arc=PI/3))
    self.play(FadeIn(line_AP_Label))
    self.wait()
    self.play(FadeOut(line_AD_m, line_AP_m), FadeIn(line_AP), line_AD_Label.animate.set_color(WHITE), line_AP_Label.animate.set_color(WHITE))
```
