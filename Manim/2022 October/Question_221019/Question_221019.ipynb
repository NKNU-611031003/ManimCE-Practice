# ManimCE

## Logo

```python
from manim import *
def ManimCELogo(self):
    # logo
    banner = ManimBanner(dark_theme=False)
    circle_banner = Circle(color="#ece6e2", fill_opacity=1).surround(banner).scale(2)
    # logo Text
    logo_Text = Text("Manim Community", font_size=48).shift(RIGHT)
    logo_MadeBy = Text("Made by", font_size=24).next_to(logo_Text, UP).shift(LEFT*2.5)
    logo_Web = Text("https://docs.manim.community/en/stable/index.html", font_size=16).next_to(logo_Text, DOWN)
    # Group
    banner_Group = VGroup(circle_banner, banner)
    logo_Text_Group = Group(logo_MadeBy, logo_Web)

    self.play(FadeIn(circle_banner), banner.create())
    self.play(banner.animate.scale(0.2), circle_banner.animate.scale(0.1))
    self.play(banner_Group.animate.next_to(logo_Text_Group, LEFT))
    self.play(FadeIn(logo_Text_Group), Write(logo_Text), run_time=2.5)
    self.wait(3)
    self.play(FadeOut(banner_Group), FadeOut(logo_Text_Group), FadeOut(logo_Text))
    self.wait()
```

## Question

```python
def main_polygon(self):
    args = parser.parse_args()

    ### Polygon ###
    square = Square(side_length=args.side_length*2)
    arc_1 = Arc(radius=args.side_length, angle=PI, arc_center=square.get_bottom())
    arc_2 = Arc(radius=args.side_length*2, angle=PI/2, arc_center=square.get_top(), start_angle=PI*3/2).shift(LEFT*args.side_length)
    polygon = VGroup(square, arc_1, arc_2)

    ### Dot and Label ###
    def Dot_and_Label(num, name, dot_pos, label_pos):
      dot[num] = Dot(point=square.get_center()).scale(args.dot_scale).shift((dot_pos)*args.side_length)
      label[num] = Tex(name, font_size=args.label_font_size).next_to(dot[num], label_pos)
      return VGroup(dot[num], label[num])
    dot = list(range(6))
    label = list(range(6))
    dot_label = list(range(6))
    dot_label[0] = Dot_and_Label(0, "$A$", LEFT+UP, (LEFT+UP)*args.label_offset_in)
    dot_label[1] = Dot_and_Label(1, "$B$", RIGHT+UP, (RIGHT+UP)*args.label_offset_in)
    dot_label[2] = Dot_and_Label(2, "$C$", RIGHT+DOWN, (RIGHT+DOWN)*args.label_offset_in)
    dot_label[3] = Dot_and_Label(3, "$D$", LEFT+DOWN, (LEFT+DOWN)*args.label_offset_in)
    dot_label[4] = Dot_and_Label(4, "$M$", DOWN, (DOWN)*0.8)
    dot_label[5] = Dot_and_Label(5, "$P$", RIGHT*0.6+DOWN*0.2, (UP)*args.label_offset_in)
    dot_group = VGroup(*[dot_label[i] for i in range(6)])

    ### Side length label ###
    def Side_label(num, label_name, dot_0, dot_1, direction):
      length_label_Text = Tex(label_name, font_size=args.label_font_size).next_to(Line(dot_0, dot_1).get_center(), direction)
      return length_label_Text
    length_label = list(range(4))
    length_label[0] = Side_label(0, "4", dot[0], dot[1], UP)
    length_label[1] = Side_label(1, "4", dot[1], dot[2], RIGHT)
    length_label[2] = Side_label(2, "2", dot[2], dot[4], DOWN)
    length_label[3] = Side_label(3, "2", dot[3], dot[4], DOWN)
    length_label_group = VGroup(*[length_label[i] for i in range(4)])

    ### All graph ###
    graphics = VGroup(polygon, dot_group, length_label_group).shift(DOWN*0.5)

    question_Text = Tex(r'求 $P$ 到 $\overline{AD}$ 邊的距離', tex_template=TexTemplateLibrary.ctex, font_size=args.font_size).next_to(graphics, UP*2)

    return polygon, dot_group, length_label_group, graphics, question_Text
```

# Run

```python
%%manim -qm -v WARNING MathQuestion_221019
import argparse
import math

parser = argparse.ArgumentParser()
parser.add_argument('-f')
parser.add_argument("--font_size", type=int, default=50)
parser.add_argument("--label_font_size", type=int, default=50)
parser.add_argument("--label_offset_in", type=float, default=0.6, help="dot label position offset with indent.")
parser.add_argument("--label_offset_out", type=float, default=1.2, help="dot label position offset with expansion.")
parser.add_argument("--side_length", type=float, default=2.0, help="Side length of half square.")
parser.add_argument("--dot_scale", type=float, default=1.0, help="Size of each point.")

class MathQuestion_221019(Scene):
  def construct(self):
    args = parser.parse_args()

    # ManimCELogo(self)
    polygon, dot_group, length_label_group, graphics, question_Text = main_polygon(self)
    self.play(FadeIn(graphics), Write(question_Text), run_time=2)
    self.wait()
```
